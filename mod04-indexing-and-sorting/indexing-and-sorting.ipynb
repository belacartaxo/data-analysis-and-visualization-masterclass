# Indexing and sorting

## Indexing

### Setting an Index
- `set_index(column_name, inplace=False)`: Sets the specified column as the DataFrame's index.
  - Example: `table.set_index("column", inplace=True)`  

- `pd.read_csv(filepath, index_col=column_name)`: Reads a CSV file and sets the specified column as the index during import.
  - Example: 
    ```python
    table = pd.read_csv('data.csv', index_col="column")
    ```

## Sorting

### Sorting by Values
- `sort_values(by, ascending=True, key=None)`: Sorts the DataFrame by one or more columns.
  - Single column: `table.sort_values("column", ascending=False)`  
  - Multiple columns: `table.sort_values(["column1", "column2"], ascending=True)`  
  - Using a custom function: `table.sort_values("column", key=lambda col: col.str.lower())`

#### Case Sensitivity in Sorting Strings

When sorting strings, the method differentiates between uppercase and lowercase letters. By default:  
- Uppercase letters are sorted before lowercase letters due to their ASCII values. 

To perform case-insensitive sorting, use the key parameter with a function like str.lower

`table.sort_values("column", key=lambda col: col.str.lower())`

### Sorting by Index
- `sort_index(inplace=False)`: Sorts the DataFrame by its index.  
  - Example: `table.sort_index(inplace=True)`

## Selecting Data

### Row Selection
- `loc`: Label-based indexing.
  - Single row as a Series: `table.loc["label"]`  
  - Single row as a DataFrame: `table.loc[["label"]]`  
  - Multiple rows: `table.loc[["label1", "label2"]]`  
  - Slicing with steps: `table.loc[start_label:end_label:step]`  

- `iloc`: Position-based indexing.
  - Specific rows by position: `table.iloc[[i, j, k]]`  
  - Slice by index range: `table.iloc[start:end]`

### Row and Column Selection
- `loc` with rows and specific columns:
  - Example: `table.loc[start_label:end_label, ['column1', 'column2']]`

### Difference Between `loc` and `iloc`

- **`loc`**:  
  - Label-based indexing.  
  - Used to select rows and columns by their labels or a boolean array.  
  - Example: `table.loc["row_label", "column_label"]`  

- **`iloc`**:  
  - Position-based indexing.  
  - Used to select rows and columns by their integer positions.  
  - Example: `table.iloc[0, 1]`  

**Key Difference**:  
- `loc` uses explicit labels from the index and column names.  
- `iloc` uses numerical indices based on the row and column order.
