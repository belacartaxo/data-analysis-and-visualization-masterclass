# Filtering

Filtering in pandas allows you to select rows based on specific conditions. When you apply a condition like `df[df.column == "label"]`, this condition returns a **Series of boolean values** (`True` or `False`) that indicate which rows satisfy the condition. This Series can then be used to filter the DataFrame and return only the rows where the condition is `True`.

## Comparison Operators
You can use standard comparison operators to filter data:
- `==`: Equals
- `!=`: Not equal
- `>`: Greater than
- `<`: Less than
- `>=`: Greater than or equal
- `<=`: Less than or equal

Example:
```python
df[df.age > 30]  # Returns rows where age is greater than 30
```

## Logical Operators
You can combine conditions using logical operators:

- &: AND (both conditions must be true)
- |: OR (either condition must be true)
- ~: NOT (negation of a condition)

Example:
```python
df[(df.age > 30) & (df.sex == "female")]  # Returns rows where age > 30 and sex is female
```

## Important functions
### .between()
The .between() method is used to filter values within a specific range. By default, it includes both endpoints. You can customize whether the range is inclusive or exclusive using the inclusive parameter.

Syntax: df[column].between(left, right, inclusive="both")
- **left**: The lower bound.
- **right**: The upper bound.
- **inclusive**: Determines if the bounds are inclusive. Options are "both", "neither", "left", or "right".

Example:
```python
df[df.age.between(20, 30)]  # Returns rows where age is between 20 and 30, inclusive of both
```

### .isin()
The .isin() method allows you to filter rows where a column's value is in a list of values.

Syntax: df[column].isin([value1, value2, ...])

Example:
```python
df[df.sex.isin(["female", "non-binary"])]  # Filters rows where sex is female or non-binary
```

### .isna() and .notna()
- .isna() returns a boolean Series indicating whether each element is NaN (Not a Number).
- .notna() returns a boolean Series indicating whether each element is not NaN.

Example:
```python
df[df.age.isna()]  # Returns rows where age is NaN
df[df.age.notna()]  # Returns rows where age is not NaN
```
