# Updating values

## Rename Columns and Index Labels

You can use the `rename` method to rename column names or index labels in a DataFrame. A `mapper` dictionary is used to specify the old and new names.

---

### Renaming Columns
To rename columns, use the `columns` argument and provide a dictionary where the keys are the old column names and the values are the new names.

#### Example:
```python
mapper = {"Regional indicator": "regional_indicator", "Ladder score": "ladder_score"}
countries.rename(columns=mapper, inplace=True)
```

### Renaming Index Labels
To rename index labels, use the `index` argument and provide a dictionary where the keys are the old index labels and the values are the new ones.

Example:
```python
mapper = {"Finland": "finland", "Denmark": "denmark"}
countries.rename(index=mapper, inplace=True)
```
The rename method allows you to update labels without modifying the data directly.


````python
import pandas as pd
houses = pd.read_csv('../data/kc_house_data.csv')
titanic = pd.read_csv('../data/titanic.csv')
netflix = pd.read_csv('../data/netflix_titles.csv', sep="|", index_col=0)
btc = pd.read_csv('../data/coin_Bitcoin.csv')
countries = pd.read_csv('../data/world-happiness-report-2021.csv')
````

````python
mapper = {"Regional indicator": "regional_indicator",	"Ladder score": "ladder_score"} 
countries.rename(columns=mapper, inplace=True)
````

````python
countries.set_index("Country name", inplace=True)
mapper = {"Finland": "finland",	"Denmark": "denmark"} 
countries.rename(index=mapper)
````

## Updating Values in a DataFrame

There are multiple ways to update values in a DataFrame, depending on whether you're replacing specific values or updating based on conditions.



### Using `replace()`

The `replace()` method allows you to substitute specific values in a column or the entire DataFrame.

#### Key Points:
- **Replace a single value:** 
  ```python
  titanic["sex"].replace("female", "F", inplace=True)
  ```

- **Replace multiple values using lists: The lists must have the same size, and the order of elements matters.**
    ```python
    titanic["sex"].replace(["female", "male"], ["F", "M"], inplace=True)
    ```

- **Replace a value with None (missing value): Wrap the value in a list, even if it's a single element.**

    ```python
    titanic["age"].replace(["?"], [None], inplace=True)
    ```
**OBS:** Show NaN values in value_counts - By default, value_counts excludes NaN. To include them, use **dropna=False**.

```python
titanic["age"].value_counts(dropna=False)
```

### Using loc
The loc method allows you to update values based on indices or boolean masks.

#### Key Points:
- **Update a single value in a specific row and column:**

    ```python
    countries.loc["Brazil", ["ladder_score"]]
    ```
- **Update multiple rows and columns:**

    ```python
    countries.loc[["Brazil", "Panama", "Honduras"], ["ladder_score"]]
    ```
- **Update multiple columns with a new value:**

    ```python
    countries.loc[["Denmark", "Sweden", "Norway"], ["regional_indicator"]] = "Scandinavia"
    ```
- **Create new columns while updating: Columns that don't exist will be created.**

    ```python
    countries.loc[["Denmark", "Sweden", "Norway"], ["regional_indicator", "ladder_score", "isCold"]] = "Scandinavia"
    ```
- **Update values using boolean masks: You can pass a condition to select specific rows and update values.**

    ```python
    houses.loc[(houses["bedrooms"] < 3), ["bedrooms"]] = 6
    ```
The loc method provides flexibility to update both existing columns and create new ones based on your conditions or indices.
