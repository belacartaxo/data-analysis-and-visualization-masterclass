# Understanding the Concept of `Figure` in Matplotlib

In **Matplotlib**, a `Figure` is the top-level container for all the elements of a visualization. It acts as a canvas where plots and other visual components (like axes, legends, titles) are drawn. Each `Figure` can contain multiple plots, but by default, each plot belongs to a single `Figure`.

## Key Points:
1. **Single Figure**: When multiple plots are drawn without creating new `Figure` instances, they are all added to the same canvas.
2. **Multiple Figures**: Creating new `Figure` instances allows you to separate plots into different canvases.

Let's explore these concepts with examples:

### Example 1: All Plots in a Single Figure
```python
import numpy as np
nums = np.linspace(0, 10, 100)

# All plots are part of the same figure
plt.plot(nums, nums)
plt.plot(nums, nums**2)
plt.plot(nums, nums**3)
```

### Example 2: Explicitly Creating a New Figure
```python
# Create a new figure before plotting
plt.figure()
plt.plot(nums, nums)
plt.plot(nums, nums**2)
plt.plot(nums, nums**3)
```

Although we explicitly called plt.figure(), all three plots are still part of the same figure because they were created after the plt.figure() call without additional figure separation.

### Example 3: Multiple Figures
```python
# Create multiple figures
plt.figure()  # First figure
plt.plot(nums, nums)

plt.figure()  # Second figure
plt.plot(nums, nums**2)
plt.plot(nums, nums**3)
```

**In this case**:
- The first Figure contains the plot y = x.
- The second Figure contains two plots: y = x^2 and y = x^3.

# Matplotlib Stylesheets and Line Customization

## Stylesheets in Matplotlib
Matplotlib provides **stylesheets** to quickly change the appearance of your plots. Stylesheets define aesthetic settings, such as colors, fonts, and line thickness, making it easy to achieve professional or theme-based designs.

### Listing Available Styles
```python
import matplotlib.pyplot as plt
print(plt.style.available)
```
This command prints a list of available styles, such as: 
`'fivethirtyeight', 'ggplot', 'seaborn', 'classic', 'dark_background'`

### Applying a Style
You can use a stylesheet by calling:

```python
plt.style.use('fivethirtyeight')
```

This applies the `fivethirtyeight` style to all subsequent plots.

### Line Customization
In addition to stylesheets, you can customize individual plot elements, such as lines and markers. Below are examples of line and marker settings:

**Example**: Customizing Lines
```python
nums = range(10)

plt.style.use('fivethirtyeight')  # Apply a style

# Plot 1: Custom line color, width, and style
plt.plot(nums, nums, color="pink", linewidth=14, linestyle="dotted")

# Plot 2: Default line style
plt.plot(nums, [x**2 for x in nums])

# Plot 3: Adding markers with customization
plt.plot(nums, [x**3 for x in nums], marker=">", markersize=20, markerfacecolor="green")

```

### Additional Line Style Customizations
#### Line Colors
You can specify colors using:
- **Named colors**: 'blue', 'green', 'red'

- **RGB tuples**: (0.1, 0.2, 0.5)

- **'-'** (solid line)  Default

Adjust with the linewidth parameter.

- **Shapes**:

### Combining Customizations
You can mix and match these customizations to create unique and visually appealing plots tailored to your data and presentation needs.

```python

## Common Attributes for `xlabel`, `ylabel`, and `title`

### 1. **`loc`**: Location of the text
- Determines the horizontal alignment of the title or label.
- Possible values:
  - `"center"` (default): Centers the text.
  - `"left"`: Aligns the text to the left.
  - `"right"`: Aligns the text to the right.
### 2. **`fontsize`**: Font size
- Adjusts the size of the text.

### 3. **`color`**: Text color
- Sets the color of the text.
- Accepts:
    - `Named colors` (e.g., "red", "blue", "green").
    - `Hexadecimal` codes (e.g., "#FF5733").
    - `RGB tuples` (e.g., (0.1, 0.2, 0.5)).
### 4. **`fontweight`**: Text weight
- Controls the thickness of the text.
- Possible values:
    - `"normal"` (default)
    - `"bold"`
    - `"light"`

### 5. **`rotation`**: Text rotation
- Rotates the text by a specified angle (in degrees).
- Common for axis labels.

### 6. **`pad`**: Padding
- Adds space between the text and the plot or axis.

````python
plt.plot(nums, nums**3)
plt.title("Plot title", loc="left", fontsize=40, color="red")
plt.xlabel("x label here")
plt.ylabel("y label here", fontsize=30, rotation="horizontal")
````

# Customizing X & Y Ticks in Matplotlib

Matplotlib provides methods to customize the ticks and their labels for both x and y axes, as well as to define the range of data displayed using axis limits.


## 1. Setting Custom Ticks: `xticks()` and `yticks()`

### **`plt.xticks() and plt.yticks()`**
- Defines the positions of the ticks on the x-axis and y-axis.
- Optional: Custom labels for each tick.
- plt.xticks(ticks, labels=None, **kwargs)

#### Syntax:
```python
plt.xticks([1, 1.7, 2, 2.3, 3, 4])  # Custom x-axis tick positions
plt.yticks([10, 20, 23, 30, 40, 60], labels=["10k", "20k", "23k", "30k", "40k", "60k"])  # Custom y-axis tick positions and labels
```
- **ticks**: A list of positions for the ticks (e.g., [1, 2, 3]).
- **labels**: A list of custom labels for the ticks (e.g., ["One", "Two", "Three"]).

## 2. Setting Axis Limits: xlim() and ylim()
### plt.xlim() and plt.ylim()
Specifies the minimum and maximum range for the x-axis and y-axis.

Syntax:
```python
plt.xlim(1, 3.5)  # Limits x-axis range to 1 to 3.5
plt.ylim(5, 50)   # Limits y-axis range to 5 to 50
```

# Adding Legends to Matplotlib Plots

Legends in Matplotlib are used to identify the data represented by different lines or markers in a plot. They enhance the interpretability of plots, especially when multiple datasets are displayed.


## Adding Legends with `plt.legend()`

The **`plt.legend()`** function is used to display a legend. Each plot must include a `label` parameter, which specifies the text to display in the legend.

### Key Parameters:
#### loc:
- Specifies the location of the legend.
- **Common values**:
    - **"best"** (default): Automatically places the legend in the best location.
    - **`"upper right`"**, **`"lower left`"**, **`"center`"**, etc.
#### shadow:
- Adds a shadow behind the legend box.
    - **Values**: True or False (default).
#### frameon:
- Determines whether the legend box has a frame.
- **Values**: True (default) or False.
#### facecolor:
- Sets the background color of the legend box.
- **Example**: "white", "lightgrey", etc.
#### fontsize:
- Controls the font size of the legend text.
- **Example**: "small", "large", 12, etc.

### Syntax:
```python
plt.plot(nums, nums, label="x")
plt.plot(nums, nums**2, label="x squared")
plt.plot(nums, nums**3, label="x cubed")
plt.legend(loc="upper right", shadow=True, frameon=True, facecolor="white")
```

````python
plt.figure()
plt.plot(nums, nums, label="x")
plt.plot(nums, nums**2, label="x squared")
plt.plot(nums, nums**3, label="x cubed")
plt.legend(loc="upper right", shadow=True, frameon=True, facecolor="white")
````

## Bar Plots 

A bar plot is a graphical representation of data where rectangular bars represent the values associated with categories. In `matplotlib`, two types of bar plots can be created: vertical and horizontal.

### 1. **Vertical Bar Plot**: 
   - In a vertical bar plot, categories are placed along the x-axis, and the values are represented by bars extending vertically. The length of each bar corresponds to the value of the category it represents.
   - You can customize the appearance of the bars by adjusting parameters such as **width** (determines the thickness of the bars) and **color** (specifies the color of the bars).

### 2. **Horizontal Bar Plot**: 
   - A horizontal bar plot is similar to a vertical bar plot but with bars extending horizontally along the y-axis. This type of plot can be useful when categories have long names or when the data values are easier to compare in a horizontal orientation.
   - The **height** parameter determines the thickness of each bar in a horizontal bar plot, while the **color** can still be adjusted as in the vertical version.

Both types of bar plots are commonly used to compare quantities across different categories, and the choice between vertical and horizontal depends on the nature of the data and the desired visualization.

### `align="edge"` in Bar Plots

In `matplotlib`, the `align` parameter is used to control the positioning of the bars in a bar plot. When creating vertical or horizontal bar plots, you can choose how the bars are aligned with respect to their positions on the axis. The `align` parameter accepts two main options: `"center"` and `"edge"`.

1. **`align="center"`** (default): 
   - This option centers the bars on their x or y positions. For example, if the category for a bar is at position `x = 2`, the bar will be centered around `x = 2`. This is the default alignment for most bar plots.
   - The width of the bar is equally distributed on both sides of the x (or y) position, which may create spacing between bars.

2. **`align="edge"`**: 
   - When `align="edge"` is used, the bar's edge will align with the specified x or y position, rather than the center. For example, if the category for a bar is at `x = 2`, the left edge of the bar will start at `x = 2`.
   - This alignment is useful when you want to create bar plots where the bars "stick" to the axis or when plotting multiple bar plots next to each other.

Choosing between `"center"` and `"edge"` depends on how you want the bars to be positioned. The `"center"` option gives a balanced look, while `"edge"` is useful when precise positioning of bars along the axis is needed or when comparing multiple datasets.

 
### Example
```python
plt.bar(plants, germinated, width=0.5, color="green")
plt.barh(plants, germinated, color="green", height=0.5)
```

````python
plants = ["Spinach", "Turnip", "Rhubarb", "Broccoli", "Kale"]
died = [10, 25, 5, 30, 21]
germinated = [74, 88, 56, 69, 59]
````

````python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, width=0.5, color="green")
plt.bar(plants, died, width=0.5, color="red", align="edge")
plt.title("Germination Rates")
plt.ylabel("Number of Seeldings")
````

````python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, width=0.5, color="green", label="germinated")
plt.bar(plants, died, width=0.5, color="red", bottom=germinated, label="died")
plt.title("Germination Rates")
plt.ylabel("Number of Seeldings")
plt.legend()
````

````python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(6,6))
plt.barh(plants, germinated, color="green", height=0.5)
plt.title("Germination Rates")
plt.ylabel("Number of Seeldings")
````

## Histograms

`plt.hist()` is used to create a histogram in `matplotlib`:

- **`np.random.randn(n)`**: This generates n random data points from a standard normal distribution (mean = 0, standard deviation = 1). These values will be visualized in the histogram.

- **`bins=n`**: The `bins` parameter specifies the number of intervals or "bins" the data will be divided into. More bins generally result in a finer resolution of the data distribution.

- **`alpha=a`**: The `alpha` parameter controls the transparency of the histogram bars. The value `a` means the bars will have a% opacity, making it possible to see overlapping data or other elements behind the bars.

```python 
plt.hist(np.random.randn(100), bins=20, alpha=0.5)
```

This code visualizes the distribution of the 100 random numbers, grouped into 20 intervals, with semi-transparent bars to make the graph more readable when layered with other elements.


````python
plt.hist(np.random.randn(100), bins=20)
````

````python
plt.hist(np.random.randn(100), bins=20, alpha=0.5)
plt.hist(np.random.randn(100), bins=20, alpha=0.5)
````

### Scatter Plots in Matplotlib

A scatter plot is used to display the relationship between two variables by plotting individual data points on a two-dimensional graph. In `matplotlib`, the `plt.scatter()` function is used to create scatter plots.

1. **Basic Usage**:
   - `plt.scatter(x, y)`: Plots data points where `x` and `y` are arrays representing the horizontal and vertical coordinates, respectively.
   - Each point represents one data observation in the dataset.

2. **Customizing Markers**:
   - You can change the appearance of the points using the `marker` parameter. This allows for different shapes to represent the data.
   - Examples of marker options:
     - `"o"`: Circle (default)
     - `"s"`: Square
     - `"^"`: Triangle (up)
     - `"v"`: Triangle (down)
     - `"*"`: Star
     - `"x"`: X-shaped marker
     - `"+""`: Plus sign

3. **Additional Customizations**:
   - **`color`**: Set the color of the points (e.g., `color="red"` or `color="#1f77b4"`).
   - **`s`**: Define the size of the markers. Pass an array to vary the size by data point.
   - **`alpha`**: Adjust transparency of the markers (e.g., `alpha=0.7` for 70% opacity).
   - **`edgecolor`**: Set the color of the marker edges.

4. **Example**:
   A scatter plot with customized markers and sizes:
   ```python
   plt.scatter(x, y, marker="^", color="blue", s=50, alpha=0.8)

### Pie Charts in Matplotlib

Pie charts are used to represent data as proportions of a whole, with each slice of the pie corresponding to a category. In `matplotlib`, the `plt.pie()` function creates pie charts.
```python
plt.pie(prices, labels = labels, autopct="%1.2f%%", explode=(0,0,0.05,0), shadow=True)
```

1. **Basic Parameters**:
   - **`x`**: The data values to represent (e.g., `prices`). The sum of these values determines the size of the pie.
   - **`labels`**: A list of category names to label each slice of the pie.
   - **`autopct`**: A format string to display the percentage value on each slice. For example:
     - `"%1.2f%%"`: Displays percentages with two decimal places.
     - `"%d%%"`: Displays percentages as whole numbers.
   - **`explode`**: A tuple specifying how far each slice is "exploded" or separated from the center. Each value corresponds to a slice; a value of `0.05` pushes that slice outward slightly.
   - **`shadow`**: If `True`, adds a shadow effect to the pie chart for a 3D look.

2. **Explaining the Example**:
   - `prices`: Represents the values for each category.
   - `labels`: Names of the categories corresponding to the `prices`.
   - `autopct="%1.2f%%"`: Displays the percentage of each slice with two decimal places.
   - `explode=(0, 0, 0.05, 0)`: The third slice is slightly separated from the pie, highlighting that category.
   - `shadow=True`: Adds a shadow effect for visual enhancement.

3. **Pie Chart Use Cases**:
   Pie charts are ideal for visualizing proportions but are most effective with a limited number of categories. Too many slices can make the chart harder to interpret.

4. **Customization**:
   - **`colors`**: Assign specific colors to slices for better differentiation.
   - **`startangle`**: Rotates the chart to start from a specific angle, often set to 90 for better orientation.
   - **`radius`**: Adjusts the size of the pie.

Pie charts are visually appealing and effective for showing percentages or proportions but should be used sparingly when there are fewer categories for clarity.

## Subplots in Matplotlib

Subplots allow multiple plots to be displayed within a single figure. They are useful for comparing data visualizations side by side. The `plt.subplot()` function is used to define individual subplots within a grid layout.

1. **Grid Layout (`plt.subplot(rows, cols, index)`)**:
   - The subplot grid is defined by the number of rows (`rows`) and columns (`cols`). Each subplot is placed at a specific position determined by the `index`.
   - For example, `plt.subplot(1, 3, 1)` creates a grid with 1 row and 3 columns, and positions the first subplot in the first column.

2. **Customizing Subplots**:
   - `plt.figure(figsize=(10, 4))`: Sets the size of the entire figure (width = 10 units, height = 4 units).
   - `plt.tight_layout()`: Automatically adjusts spacing between subplots to prevent overlap between axes labels, titles, or other elements.

3. **Example Explanation**:
   - The example creates three subplots:
     - First subplot: Plots a linear relationship (`nums` vs. `nums`).
     - Second subplot: Plots a quadratic relationship (`nums` vs. `nums**2`) in red.
     - Third subplot: Plots a cubic relationship (`nums` vs. `nums**3`) in green.

### Sharing Axes with `sharex` and `sharey`

When creating subplots, its often useful to ensure they share the same axis scales for better comparison. This is controlled using the `sharex` and `sharey` parameters in the `plt.subplots()` function:

1. **`sharex=True`**:
   - Ensures all subplots share the same x-axis scale. Any zooming or panning applied to one subplot's x-axis will reflect in all subplots.

2. **`sharey=True`**:
   - Ensures all subplots share the same y-axis scale, enabling direct comparison of values across subplots.

3. **Usage with Subplots**:
   Instead of using `plt.subplot()`, you can use `plt.subplots()` to define subplots in a more flexible way:
   ```python
   fig, axes = plt.subplots(1, 3, figsize=(10, 4), sharex=True, sharey=True)
   axes[0].plot(nums, nums)
   axes[1].plot(nums, nums**2, color="red")
   axes[2].plot(nums, nums**3, color="green")
   plt.tight_layout()


````python
nums = np.arange(5)
plt.figure(figsize=(10,4))
plt.subplot(1, 3, 1)
plt.plot(nums, nums)
plt.subplot(1, 3, 2)
plt.plot(nums, nums**2, color="red")
plt.subplot(1, 3, 3)
plt.plot(nums, nums**3, color="green")
plt.tight_layout()
````
