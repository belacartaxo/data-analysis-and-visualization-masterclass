# Seaborn

## 1. **Loading and Setting Themes**
- **`sns.load_dataset()`**: Loads a dataset provided by Seaborn (e.g., `tips`, `flights`). This is useful for quickly accessing preloaded datasets for visualization or testing.
- **`sns.set_theme()`**: Applies a consistent theme to all visualizations, improving aesthetics. Use it to set the overall style of your plots.

---

## 2. **Scatter Plots**
- **`sns.scatterplot(data, x, y)`**: Creates a scatter plot of two variables. Useful for analyzing relationships between two continuous variables.
  - **`hue`**: Differentiates points by color based on a categorical variable.
  - **`style`**: Differentiates points by marker style based on a categorical variable.
  - **`size`**: Adjusts the size of points based on a numerical variable.
  - **When to use**: Compare trends, clusters, or outliers in paired data.

---

## 3. **Line Plots**
- **`sns.lineplot(data, x, y)`**: Creates a line plot to visualize trends over a continuous variable like time.
  - **`estimator`**: Determines how data is aggregated (e.g., mean, median).
  - **`hue`**: Adds a color differentiation by a categorical variable.
  - **When to use**: Ideal for time series analysis or visualizing trends.

---

## 4. **Relational Plots**
- **`sns.relplot()`**: A versatile function for creating scatter or line plots with faceting options.
  - **`col` and `row`**: Splits the visualization into multiple subplots based on categorical variables.
  - **`kind`**: Defines the type of plot (`"scatter"` or `"line"`).
  - **`height`** and **`aspect`**: Control the size and aspect ratio of subplots.
  - **When to use**: For comparing relationships across subsets of data.

---

## 5. **Histograms**
- **`sns.histplot()`**: Plots the distribution of a single variable or compares distributions.
  - **`hue`**: Splits the distribution by a categorical variable.
  - **`multiple`**: Defines how distributions are layered (`"dodge"`, `"stack"`, `"fill"`).
  - **`bins`**: Specifies the number of bins for the histogram.
  - **`kde`**: Adds a kernel density estimate overlay.
  - **When to use**: To analyze the frequency distribution of a variable.

---

## 6. **Kernel Density Estimation (KDE)**
- **`sns.kdeplot()`**: Estimates the probability density function of a variable.
  - **`hue`**: Differentiates distributions by a categorical variable.
  - **`x` and `y`**: Visualizes joint distributions in two dimensions.
  - **When to use**: To smooth out data distribution and identify patterns.

---

## 7. **Rug Plots**
- **`sns.rugplot()`**: Adds a "rug" of tick marks along an axis to represent data points.
  - **`height`**: Controls the height of the ticks.
  - **`alpha`**: Adjusts transparency.
  - **When to use**: To add detail about exact data values in addition to other visualizations.

---

## 8. **Distributions**
- **`sns.displot()`**: Combines histogram, KDE, and rug plot functionality.
  - **`kind`**: Specifies the plot type (`"hist"`, `"kde"`).
  - **`height`** and **`aspect`**: Adjust the size and proportion of the figure.
  - **`col`**: Creates multiple plots for subsets of data.
  - **When to use**: For exploratory data analysis across subsets of data.

---

### **How to Analyze:**
1. **Understand your data**: Know whether you are working with categorical, numerical, or time-series data.
2. **Choose the function**:
   - Use scatter plots for relationships.
   - Use line plots for trends.
   - Use histograms and KDE for distributions.
3. **Leverage attributes**: Use options like `hue`, `col`, `row`, or `style` to provide meaningful comparisons.
4. **Iterate**: Experiment with plot types and options to uncover patterns, anomalies, or insights.


```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.set_theme()
sns.scatterplot(data=tips, x="total_bill", y="tip")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", style="time")
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", size="size")
```

```python
flights = sns.load_dataset("flights")
```

```python
sns.lineplot(data=flights, x="year", y="passengers")
```

```python
sns.lineplot(data=flights, x="year", y="passengers", estimator="median")
```

```python
sns.lineplot(data=flights, x="year", y="passengers", hue="month")
```

```python
trips = sns.load_dataset("taxis", parse_dates=["pickup", "dropoff"])
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", col="sex", hue="smoker", row="time")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="line", col="sex", hue="smoker", row="time")
```

```python
sns.relplot(data=tips, x="total_bill", y="tip", kind="line", col="sex", hue="smoker", row="time", height=3, aspect=1.5)
```

```python
sns.histplot(data=tips, x="tip")
```

```python
sns.histplot(data=tips, x="tip", hue="time", multiple="dodge", bins=8, kde=True)
```

```python
sns.kdeplot(data=tips, x="tip", hue="time")
```

```python
sns.histplot(data=tips, x="tip", y="total_bill")
```

```python
sns.kdeplot(data=tips, x="tip", y="total_bill")
```

```python
sns.rugplot(data=tips, x="total_bill", height=0.2)
```

```python
sns.kdeplot(data=tips, x="total_bill")
sns.rugplot(data=tips, x="total_bill", height=0.03, alpha=0.5)
```

```python
sns.displot(kind="hist", data=tips, x="total_bill", height=5, aspect=1.5, col="time", rug=True)
```
