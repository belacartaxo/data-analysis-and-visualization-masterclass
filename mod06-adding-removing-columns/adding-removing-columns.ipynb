# Adding and Removing Columns

## Dropping Columns

To remove columns from a DataFrame, you can use the `drop()` method.

### Syntax
- `df.drop(labels, axis=1, inplace=False)`
- `labels`: Specifies the column(s) to drop. You can pass a single column name or a list of column names.
- `axis=1`: Indicates that you're dropping columns (not rows, which would be `axis=0`).
- `inplace=False`: If set to `True`, the operation modifies the DataFrame in place. If `False`, it returns a new DataFrame without the dropped columns.

### Example:

1. **Dropping a single column**:
    ```python
    btc.drop(labels="Symbol", axis=1, inplace=True)  # Removes the "Symbol" column from btc
    ```

2. **Dropping multiple columns**:
    ```python
    btc.drop(labels=["column1", "column2"], axis=1, inplace=True)  # Removes multiple columns
    ```
3. **Alternative syntax (using columns)**:
    ```python
    btc.drop(columns=["column1", "column2"])  # Removes the specified columns
    ```

## Dropping Rows

To remove rows from a DataFrame, you can use the `drop()` method.

### Syntax:
- `df.drop(labels, axis=0, inplace=False)`
  - `labels`: Specifies the row(s) to drop. You can pass a single label or a list of labels.
  - `axis=0`: This is the default axis, so it doesn't need to be specified when dropping rows.
  - `inplace=False`: If set to `True`, the operation modifies the DataFrame in place. If `False`, it returns a new DataFrame without the dropped rows.

### Examples:

1. **Dropping a row by label**:
   ```python
   countries.drop(labels="Denmark")  # Removes the row with the label "Denmark"
    ```
2. **Dropping a row by default index**:
    ```python
    btc.drop(0)  # Removes the row at index 0
    ```
3. **Dropping rows using a range of indices (when the index is not numeric)**:
    ```python
    countries.drop(countries.index[0:3])  # Removes rows with indices 0, 1, and 2
    ```

````python
countries.set_index("Country name", inplace=True)

countries.drop(labels="Denmark", axis=0)
````

## Adding Static Columns

You can add new columns to a DataFrame in different ways, either by appending a static value or inserting a column at a specific position.

### 1. **Adding a column at the end of the DataFrame:**

You can assign a value or a Series to a new column in the DataFrame. If you assign a single value, it will be applied to all rows.

**Example:**
```python
titanic["humans"] = "Human"  # Adds a new column "humans" with the value "Human" for all rows
```

### 2. **Inserting a column at a specific position**:
You can use the insert() method to add a column at a specific position in the DataFrame. The first argument specifies the index where the column should be inserted, and the second argument is the column name. The third argument is the value to fill the column, which can be a single value or a Series.

**Example**:

```python
titanic.insert(0, "specie", "T-A")  # Adds a new column "specie" at the first position with the value "T-A"
```
### 3. **Adding a column with varied values**:
Instead of a static value, you can also pass a Series with different values to a new column.

**Example**:

```python
titanic["age_group"] = [22, 34, 28, 41, 56]  # Adds a new column "age_group" with different values for each row
```
