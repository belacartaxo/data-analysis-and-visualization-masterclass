# Seaborn categorical plots

## 1. **Count Plots**
- **`sns.countplot()`**: Displays the frequency of categories in a dataset.
  - **Attributes**:
    - **`x` or `y`**: Specify the category variable (horizontal or vertical bar plot).
    - **`hue`**: Differentiates categories within each bar using color.
  - **When to use**: To analyze how often a specific category appears (e.g., count of days, distribution of time across days).

---

## 2. **Swarm and Strip Plots**
- **`sns.swarmplot()`**: Creates a scatter plot for categorical data with points adjusted to avoid overlap.
  - **When to use**: To visualize data distribution with individual data points while maintaining category grouping.
- **`sns.stripplot()`**: Similar to `swarmplot`, but without adjusting for overlaps, resulting in points stacked on top of each other.
  - **When to use**: For quick insights into data spread and clustering.

---

## 3. **Box and Boxen Plots**
- **`sns.boxplot()`**: Displays data distribution based on quartiles with outliers.
  - **Attributes**:
    - **`x` and `y`**: Specify the categorical and numerical variables.
  - **When to use**: To visualize data spread, detect outliers, and compare distributions across categories.
- **`sns.boxenplot()`**: Similar to `boxplot` but optimized for large datasets by showing additional granularity.
  - **When to use**: For datasets with many observations, where finer details of the distribution are necessary.

---

## 4. **Violin Plots**
- **`sns.violinplot()`**: Combines boxplot and KDE to show data distribution and density.
  - **Attributes**:
    - **`hue`**: Differentiates distributions within each category.
    - **`split`**: Splits the violin plot into two halves for comparing distributions.
  - **When to use**: To visualize distribution and density across categories, especially when understanding nuances is crucial.

---

## 5. **Bar Plots**
- **`sns.barplot()`**: Shows the average value for each category with confidence intervals.
  - **Attributes**:
    - **`x` and `y`**: Specify the categorical and numerical variables.
  - **When to use**: To summarize data by categories, comparing mean or aggregated values.

---

## 6. **Categorical Plots**
- **`sns.catplot()`**: A versatile function that unifies different categorical plots (e.g., boxplot, violinplot, barplot).
  - **Attributes**:
    - **`kind`**: Specifies the type of plot (`"strip"`, `"swarm"`, `"box"`, `"violin"`, `"bar"`, etc.).
    - **`col`** and **`row`**: Create multiple subplots for faceted analysis.
    - **`hue`**: Differentiates data within each category by color.
    - **`height`** and **`aspect`**: Adjust figure size and aspect ratio.
  - **When to use**: For flexible categorical data visualizations, especially when you want to compare data across multiple facets.

---

### **How to Analyze**:
1. **Understand your data**: Identify if you're working with categorical variables (e.g., boroughs, days) and numerical variables (e.g., total, distance).
2. **Choose the plot**:
   - Use **count plots** for category frequency.
   - Use **box/boxen/violin plots** for distribution analysis.
   - Use **swarm/strip plots** for visualizing individual data points.
   - Use **bar plots** for aggregated summaries.
3. **Refine with attributes**: Add hue, split, or facets (with `catplot`) for deeper comparisons.
4. **Explore patterns**: Look for trends, outliers, overlaps, and differences between categories.


```python
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset("tips")
tips
```

```python
sns.set_theme()
sns.countplot(data=tips, x="day", hue="time")
```

```python
sns.countplot(data=tips, y="day", hue="time")
```

```python
trips = sns.load_dataset("taxis")
trips
```

```python
sns.stripplot(data=trips, x="pickup_borough", y="distance")
```

```python
sns.swarmplot(data=trips.nlargest(600, "total"), x="pickup_borough", y="distance")
```

```python
sns.boxplot(data=trips, x="pickup_borough", y="total")
```

```python
sns.boxenplot(data=trips, x="pickup_borough", y="total")
```

```python
sns.violinplot(data=trips, x="pickup_borough", y="total")
```

```python
sns.violinplot(data=trips, x="pickup_borough", y="total", hue="color")
```

```python
sns.violinplot(data=trips, x="pickup_borough", y="total", hue="color", split=True)
```

```python
sns.barplot(data=trips, x="pickup_borough", y="distance")
```
